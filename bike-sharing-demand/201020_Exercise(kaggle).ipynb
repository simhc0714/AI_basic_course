### 가장 간단한 모델 만들기
* 머신에게 입력데이터와 출력이 나온다.
    * 분리해서 알려준다(학습시킨다) X_train, y_train
    * 머신러닝 모델을 선택하고
    * 머신러닝 모델에 데이터를 넣어서 학습
    * 학습시킨 모델이 새로운 데이터(x_test)을 넣어서 출력예측(y_test)
    * sub count에 y_test를 써서 제출한다.

```python
sel = ['season', 'weather']
X_train = bike_train[sel] # input
y_train = bike_train['count'] # output


X_test = bike_test[sel]
```

```python
# Model Selection (LinearRegression)
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # model selection

# model.fit(input data, output data)
model.fit(X_train, y_train)
```

```python
print(model.coef_) # 회귀 계수(기울기)
print(model.intercept_) # 절편
```

* y(렌탈대수) = 26.71 * x1(계절) + (-37.18 * x2(날씨)) + 177.36

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np


fig = plt.figure(figsize=[10,8])
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-500, 500, 100)
Y = np.arange(-500, 500, 100)
X, Y = np.meshgrid(X, Y)
Z = 26.71*X - 37.18*Y + 177.36

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
