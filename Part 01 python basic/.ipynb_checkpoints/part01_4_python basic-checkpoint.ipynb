## 2) numpy와 matplotlib 모듈 개요

```python
import numpy as np # 선형대수
import matplotlib.pyplot as plt
```

```python
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
# 기본값이 50으로 설정되어있음.
x = np.linspace(0, 14, 100) # 0부터 14까지 동일한 간격의 값 100개 만들기
x
```

```python
type(x)
```

```python
y = np.sin(x)
y
```

```python
# matplotlib 모듈에서 불러온 plt 메서드로 x, y를 시각화 해준다.
plt.plot(x,y)
```

```python
# 여러개의 그래프를 그려보자!

y1 = np.sin(x)
y2 = np.sin(x)*2
y3 = np.sin(x)/4

plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(x)*2")
plt.plot(x, y3, label="sin(x)/4")
plt.legend() # 범례
plt.title("sin plot") # 제목
```
