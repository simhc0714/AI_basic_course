### matplotlib의 기능
 * plt.legend() : 범례를 표시 (label)이 있어야.

 ### seaborn
  * sns.load_dataset() : 데이터 셋 로드. github 제공가능한 데이터 리스트 있음.
  
 ### pandas 로 가져와서.
  * haed(), tail() : 위 아래 기본 5개 가져오기
  * info() : 데이터의 정보(행이름, 열의 자료형, 데이터 수 등)
  * describe() : 수치형 데이터에 대해서, 몇개의 데이터가 있는지, 평균, 표준편차, 사분위수, 최소, 최대값 등
  * columns : 열이름 확인하기
  * isnull().sum() : 결측치 개수 확인

```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(8,6)) # 단위: 인치
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("total bills versus days")
plt.show()
```

```python
# hue 라는 속성으로 데이터를 추가할 수 있다.

plt.figure(figsize=(8,6)) # 단위: 인치
sns.barplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.title("total bills versus days")
plt.show()
```

### 회귀분석. Regression

```python
plt.plot(tips.total_bill, tips.tip, 'o') # data x, data y, label mark
```

```python
# 회귀선 그리기
sns.set_style=('darkgrid')
sns.lmplot(x="tip", y="total_bill", data=tips, size=7)
plt.show()
```

```python
# 회귀선 그리기
sns.set_style=('darkgrid')
sns.lmplot(x="tip", y="total_bill", hue='smoker', data=tips, size=7)
plt.show()
```

### 03 항공 데이터 시각화

```python
fg = sns.load_dataset('flights')
fg.head() # 년-월별 승객 수
```

```python
fg.info()
```

```python
plt.figure(figsize=(11,6)) # 단위: 인치
sns.barplot(x='year', y='passengers', data=fg)
plt.title("passengers versus month")
plt.show()
```

```python
# 1960년 최대값과 1950년 최대값이 같을까
fg.loc[fg['year'] == 1960]
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='year', y='passengers', data=fg)
```
