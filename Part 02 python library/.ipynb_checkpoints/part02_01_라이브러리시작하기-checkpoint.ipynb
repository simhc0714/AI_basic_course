## 시각화 라이브러리
- matplotlib (pyplot,,,)
- seaborn

```python
import matplotlib.pyplot as plt
```

```python
# 행렬 만들기
x = np.linspace(0, 14, 100) # linespace(시작값, 끝값, 개수)
x
```

```python
y = np.sin(x) # 100개의 x값에 대하여 100개의 sin 값을 구해서 y에 대입.
plt.plot(x, y) # draw graph
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기
plt.plot(x, y1, x, y2, x, y3)
```

```python
# 제목 달기
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'sin(x) * 2')
plt.plot(x, y3, label = 'sin(x) * 4')
plt.legend() # 범례
plt.title("sin plot") #제목
```

```python
# 실습 3-1 cosine 그래프 그려보기
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4

# 제목 달기
plt.plot(x, y1, 'o', label = 'cos(x)')
plt.plot(x, y2, 'x', label = 'cos(x) * 2')
plt.plot(x, y3, '^', label = 'cos(x) * 4')
plt.legend() # 범례
plt.title("cosine plot") #제목
```

### 데이터를 불러와서 이에 대해서 간단한 시각화를 통해 알아보기

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips") # pandas의 데이터프레임 형태로 가져오게 됨.
tips
```

```python
type(tips)
# pandas 데이터는 series와 frame형태를 갖는다. series는 1차형, dataframe은 2차형
```

* [데이터].info(): 데이터 정보
* [데이터].describe() : 수치형 데이터 요약값
* [데이터].shapee() : row, column information
* [데이터].columns() : 컬럼명, 제목
* [데이터].head() : 앞에서 기본적인 데이터 몇개(5) 확인해준다.
* [데이터].tail() : 뒤에서 기본적인 데이터 몇개(5) 확인해준다.
* [데이터].isnull().sum() : 각 열의 결측치(비어 있는 값)의 개수.
