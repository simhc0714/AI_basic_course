### 데이터 선택하기
- [데이터].loc[행, 열]
- [데이터].loc[조건, 열]
- [데이터][조건]

```python
# A1 : 75-100 3151 ~
# A2 : 50-75  2127(not includede) ~ 3151 (included)
# A3 : 25~50  1462 ~ 2127
# A4 : 0~25   ~ 1462
```

```python
train.describe()
```

```python
# A1 : 75~100  3151 ~
# A2 : 50~75   2127(포함안함) ~ 3151(포함)
# A3 : 25~50   1462 ~ 2127
# A4 : 0~25    ~1462
tmp_A1 = train[train['total_rooms'] > 3151]
tmp_A2 = train[(train['total_rooms'] > 2127) & (train['total_rooms'] <= 3151)]
tmp_A3 = train[(train['total_rooms'] > 1462) & (train['total_rooms'] <= 2127)]
tmp_A4 = train[train['total_rooms'] <= 1462]

print(len(tmp_A1) + len(tmp_A2) + len(tmp_A3) + len(tmp_A4) )
```

```python
### 새로운 컬럼 'room_level'

train.loc[train['total_rooms'] > 3151, 'room_level'] = 1
train.loc[(train['total_rooms'] > 2127) & (train['total_rooms'] <= 3151), 'room_level'] = 2
train.loc[(train['total_rooms'] > 1462) & (train['total_rooms'] <= 2127), 'room_level'] = 3
train.loc[train['total_rooms'] <= 1462, 'room_level'] = 4
train
```

```python
# 나이대별 그룹 분류 'housing_median_age'
# A1 : 75~100  37 ~
# A2 : 50~75   29(포함안함) ~ 37(포함)
# A3 : 25~50   18 ~ 29
# A4 : 0~25    ~ 18

tmp_G1 = train[train['housing_median_age'] > 37]
tmp_G2 = train[(train['housing_median_age'] > 29) & (train['housing_median_age'] <= 37)]
tmp_G3 = train[(train['housing_median_age'] > 18) & (train['housing_median_age'] <= 29)]
tmp_G4 = train[train['housing_median_age'] <= 18]

print(len(tmp_G1) + len(tmp_G2) + len(tmp_G3) + len(tmp_G4) )
```
