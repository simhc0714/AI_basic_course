# 새 섹션

```
### 데이터 읽기
text = open("prince_ko.txt").read()
text
```

```
import numpy as np
prince_mask = np.array(Image.open("prince2.png"))
prince_mask[0]   # 이미지 배열화된 내용 하나의 정보 확인
```

```
from konlpy.tag import Okt  

t = Okt()
doc_nouns = t.nouns(text)
doc_nouns
```

```
ko = nltk.Text(text, name="왕자")
print(len(ko.tokens))
```

```
data = ko.vocab().most_common(3000)
```

```
### 워드 클라우드 표현을 위한 데이터 생성
### 약간의 시간이 필요.
wc = WordCloud(background_color='white', 
               max_words=500,
               mask=prince_mask,   
               contour_width=3, 
               contour_color="steelblue", 
               font_path=path).generate_from_frequencies(dict(data))
```

```
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
