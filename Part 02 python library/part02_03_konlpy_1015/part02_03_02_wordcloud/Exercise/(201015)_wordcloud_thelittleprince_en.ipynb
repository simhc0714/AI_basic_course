```
# 그래프를 그려보자. 이번에는 정상
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('seaborn-pastel')
plt.show()
```

# 새 섹션

```
### 데이터 읽기
text = open("The_Little_Prince.txt").read()
text
```

```
import numpy as np
prince_mask = np.array(Image.open("prince2.png"))
prince_mask[0]   # 이미지 배열화된 내용 하나의 정보 확인
```

```
### 워드 클라우드 표현을 위한 데이터 생성
wc = WordCloud( background_color='white',   # 배경색
               max_words=500,              # 최대 표시 단어
               mask=prince_mask,             # 마스크 이미지
               contour_width=3,             # 외곽선 
               contour_color="steelblue" )  # 색
wc.generate(text)
wc.words_
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))  # 크기
plt.imshow(prince_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```
plt.figure(figsize=(15,8))
plt.imshow(wc, interpolation="bilinear") 
plt.axis("off")
plt.show()
```
